# Degree Scaling Study

Loads `results/scaling/scaling_table.csv` produced by:

```bash
py -3.13 experiments/sweep.py scaling
# or full protocol:
py -3.13 experiments/sweep.py scaling  # uses experiments/configs/default.json
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

root = Path("..") if Path("../results").exists() else Path(".")
csv_path = root / "results" / "scaling" / "scaling_table.csv"

if not csv_path.exists():
    raise FileNotFoundError(
        f"Run experiments/sweep.py scaling first (missing {csv_path})"
    )

df = pd.read_csv(csv_path)
display(df)

fig, ax = plt.subplots(1, 2, figsize=(10, 3.5))
ax[0].plot(df["degree"], df["train_mse"], "o-", label="learned")
ax[0].plot(df["degree"], df["analytic_train_mse_flat_circuit"], "s--", label="PL analytic (flat circuit)")
ax[0].set_yscale("log")
ax[0].set_xlabel("Degree")
ax[0].set_ylabel("Train MSE")
ax[0].legend()
ax[0].set_title("Scaling cliff (seed 0)")

ax[1].bar(df["degree"].astype(str), df["train_time_s"])
ax[1].set_xlabel("Degree")
ax[1].set_ylabel("Train time (s)")
ax[1].set_title("Wall-clock training time")
plt.tight_layout()
plt.show()
````
