# Baseline Comparison (T1, degree 5)

Loads `results/paper/baseline_comparison_d5.csv` produced by:

```bash
py -3.13 -m experiments.summarize baseline
```

The CSV includes three rows: gradient Adam, PennyLane `poly_to_angles`, and Chao/pyqsp Laurent completion.
Flat-circuit MSE for analytic rows reflects convention mismatch; see `pyqsp_reconstruction_max_error` for native Chao accuracy.

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import json

root = Path("..") if Path("../results").exists() else Path(".")
csv_path = root / "results" / "paper" / "baseline_comparison_d5.csv"
curve_path = root / "results" / "paper" / "loss_curve_d5_seed0.json"

if not csv_path.exists():
    raise FileNotFoundError(
        f"Run experiments/summarize.py baseline first (missing {csv_path})"
    )

df = pd.read_csv(csv_path)
display(df)

curve = json.loads(curve_path.read_text(encoding="utf-8"))
plt.figure(figsize=(6, 3))
plt.semilogy(curve["steps"], curve["loss"])
plt.xlabel("Step")
plt.ylabel("MSE")
plt.title("Training loss (seed 0, degree 5)")
plt.tight_layout()
plt.show()
````
